### Fazer a leitura do arquivo disponibilizado via HTML, e salvar todos os dados num arquivo CSV

```python
# Importação do pacote Beautiful e do CSV
from bs4 import BeautifulSoup
import csv
```

```python
# Abrindo o arquivo HTML disponibilizad
with open('fortune.html.html', 'r', encoding = 'utf8') as html_file:
  pagina = BeautifulSoup(html_file, 'html.parser')
```

```python
# Verificando o conteúdo extraído e ocultando
# pagina
```

```python
# Localizando a tabela pela classe
tabela = pagina.find('div', {'class': 'rt-table'})
```

```python
# Consultando a tabela
# tabela
```

```python
# Puxando as linhas
linhas = tabela.find('div', {'class': 'rt-tbody'}).find_all('div', {'class': 'rt-tr-group'})
```

```python
# linhas
```

```python
dados = []

# Garantir que apenas as 100 primeiras linhas sejam consideradas
for linha in linhas[:100]:
    # Encontrando os dados através da tag div na classe search...
    colunas = linha.find_all('div', {'class': 'searchResults__cellContent--3WEWj'})

    # Extrair o texto de cada coluna e remover espaços adicionais
    linha_extraida = [coluna.text.strip() for coluna in colunas]

    # Adicionar a linha aos dados
    dados.append(linha_extraida)
```

```python
# Validando os dados adicionados --> Apenas o primeiro valor
dados[0]
```

```python
# Definindo os cabeçalhos
header = [
  'rank',
  'name',
  'revenues',
  'revenues-percent-change',
  'profits',
  'profits-percent-change',
  'assets',
  'market-value',
  'employees'
]
```

```python
# Escrevendo os dados no arquivo CSV
with open('fortune500.csv', 'w', encoding = 'utf8', newline = '') as csv_file:
  writer = csv.writer(csv_file, delimiter= ',')
  writer.writerow(header)
  writer.writerows(dados)

print('Arquivo fortune.csv criado com sucesso!')
```

```python
# Importando o pandas
import pandas as pd

# Lendo o arquivo gerado
fortune_def = pd.read_csv('fortune500.csv', delimiter= ',')
```

```python
# Criando uma função para limpeza de caracteres
def clean_column(valor):
  if isinstance(valor, str):
    valor = valor.replace('$','').replace(',', '').replace('%', '').strip()
  return valor
```

```python
# Determinando quais colunas devem ser limpas
columns_to_clean = [
    'revenues',
    'revenues-percent-change',
    'profits',
    'profits-percent-change',
    'assets',
    'market-value'
]
```

```python
# Aplicando a função
for column in columns_to_clean:
  fortune_def[column] = fortune_def[column].apply(clean_column)
```

```python
# Checando as mudanças após limpeza
fortune_def.head(5)
```

```python
# Salvando o DataFrame limpo
fortune_def.to_csv('fortune100_limpo.csv', index = False, sep = ',')
print("Arquivo 'fortune100_limpo.csv' gerado com sucesso")
```
